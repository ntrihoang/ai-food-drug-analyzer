# Tiền xử lí data từ drugbank

```python
import pandas as pd
import sqlite3
df = pd.read_csv("../data/raw_data/clean_drug_interactions_with_food.csv")
conn = sqlite3.connect("../data/database/drug-food-interaction.db")
```

```python
df.head(10)
```

```python
def clean_text(text):
    text = str(text).lower().strip()
    return text

df["drug_name"] = df["drug_name"].apply(clean_text)
df["food_name"] = df["food_name"].apply(clean_text)

# description,source chỉ cần bỏ khoảng trắng 
df["description"] = df["description"].apply(lambda x: str(x).strip()) 
df["source"] = df["source"].apply(lambda x: str(x).strip())

df.head()
```

```python
# tách food_name về dạng chuẩn (unique)

df["food_name"] = df["food_name"].str.split(";")

df = df.explode("food_name")

df["food_name"] = df["food_name"].str.strip()
df.head()
```

```python
df.to_csv("../data/clean_data/clean_drugbank_data.csv",index=False)
```
