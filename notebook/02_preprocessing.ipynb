# Tiền xử lí dữ liệu

```python
import pandas as pd
import re

df = pd.read_csv('../data/raw_data/ddinter2_dfi_data.csv')

df.head()
```

```python
df.info()
```

## Kiểm tra rỗng và lặp

```python
# kiểm tra dòng rỗng
df.isnull().sum()
```

```python
# Kiểm tra trùng lặp
df.duplicated().sum()
```

## Chuẩn hóa text


```python
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]','',text) #bỏ kí tự không phải chữ/số/khoảng trắng
    text = re.sub(r'\s+',' ',text)
    return text

df["drug_name"] = df["drug_name"].apply(clean_text)
df["food_name"] = df["food_name"].apply(clean_text)

# description,source chỉ cần bỏ khoảng trắng 
df["description"] = df["description"].apply(lambda x: str(x).strip()) 
df["source"] = df["source"].apply(lambda x: str(x).strip())

df.head()
```

## Chuẩn hóa cột severity

```python
def normalize_severity(x):
    x = str(x).lower()

    if "major" in x:
        return "High"
    elif "minor" in x:
        return "Low"
    else:
        return "Moderate"
    
df['severity'] = df['severity'].apply(normalize_severity)

print(df["severity"].value_counts())
```
